```python
from ultralytics import YOLO
import os
model_p = os.path.join('yolov8s_playing_cards.pt')

model = YOLO(model_p)
results = model.predict('1.jpg')
```

```python
# draw bounding boxes on image

results[0].show()
print(results[0].boxes.data.tolist())
```

```python
# display the output video using cv2
import cv2
import os
from ultralytics import YOLO



model = YOLO(os.path.join('yolov8s_playing_cards.pt'))
cap = cv2.VideoCapture()
cap.open(os.path.join('videos', 'video.mp4'))


while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    results = model(frame)[0]
    for result in results.boxes.data.tolist():
        x1, y1, x2, y2 = list(map(int, result[:4]))
        confidence = result[4]
        print(result)
       
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
    cv2.imshow('frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    
cap.release()
cv2.destroyAllWindows()
```
